```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
x_actul=[1,0,0,0,1,1,0,1,1,0]
x_pred =[0,0,1,1,1,1,0,1,1,0]
```

```python
confusionmatrix=confusion_matrix(x_actul,x_pred)
print(confusionmatrix)
```

```python
accuracy=accuracy_score(x_actul,x_pred)
```

```python
print(accuracy)
```

```python
precisionscore=precision_score(x_actul,x_pred)
print(precisionscore)
```

```python
recall=recall_score(x_actul,x_pred)
print(recall)
```

```python
f1score=f1_score(x_actul,x_pred)
print(f1score)
```
